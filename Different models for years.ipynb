```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy import stats
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn import metrics
import scipy
from sklearn import preprocessing
```

##### 2015

```python
fmli = pd.read_csv("fmli_15_clean.csv")
```

```python
fmli = fmli[(fmli["PSU"] == "S49B")]
```

```python
fmli.drop(["PSU"], axis = 1, inplace = True)
```

```python
log = np.log(fmli)
log["OWN"] = fmli["OWN"]
log["DEGREE"] = fmli["DEGREE"]
log["VEHICLES"] = fmli["VEHICLES"]
```

```python
sf2015msa = pd.read_csv("census_sf_msa_2015_cleaned.csv")
sf2015 = pd.read_csv("census.tracts.sf.2015.cleaned.csv")
```

```python
dm1 = sf2015[["CT.SIZE", "CT.AVGINCOME", "CT.ROOMS", "CT.OWN", "CT.DEGREE", "CT.VEHICLES"]]
dm2 = sf2015msa[["SFMSA.SIZE", "SFMSA.AVGINCOME", "SFMSA.ROOMS", "SFMSA.OWN", "SFMSA.DEGREE", "SFMSA.VEHICLES"]]
```

```python
log = np.log(fmli)
log["OWN"] = fmli["OWN"]
log["DEGREE"] = fmli["DEGREE"]
log["VEHICLES"] = fmli["VEHICLES"]
```

```python
log_dm1 = np.log(dm1)
log_dm1["CT.OWN"] = dm1["CT.OWN"]
log_dm1["CT.DEGREE"] = dm1["CT.DEGREE"]
log_dm1["CT.VEHICLES"] = dm1["CT.VEHICLES"]
log_dm1["CT.Geo_GEOID"] = sf2015["CT.Geo_GEOID"]
log_dm1["CT.Geo_NAME"] = sf2015["CT.Geo_NAME"]
```

```python
log_dm2 = np.log(dm2)
log_dm2["SFMSA.OWN"] = dm2["SFMSA.OWN"]
log_dm2["SFMSA.DEGREE"] = dm2["SFMSA.DEGREE"]
log_dm2["SFMSA.VEHICLES"] = dm2["SFMSA.VEHICLES"]
log_dm2["SFMSA.Geo_GEOID"] = sf2015msa["SFMSA.Geo_GEOID"]
log_dm2["SFMSA.Geo_NAME"] = sf2015msa["SFMSA.Geo_NAME"]
```

```python
pred_dm1 = pd.DataFrame()

for i in range(18, len(log.columns)):
    df = log[["HHSIZE", "INCOME", "ROOMS", "OWN", "DEGREE", "VEHICLES", log.columns[i]]]
    df.replace(np.Inf, np.nan, inplace = True)
    df.replace(-np.Inf, np.nan, inplace = True)
    df.dropna(inplace = True)
    df.reset_index(inplace = True)
    X = df[["HHSIZE", "INCOME", "ROOMS", "OWN", "DEGREE", "VEHICLES"]]
    Y = df[log.columns[i]]
    X1 = sm.add_constant(X)
    model = linear_model.Lasso(alpha=0.03).fit(X1, Y)
    log_dm1.replace(np.Inf, np.nan, inplace = True)
    log_dm1.replace(-np.Inf, np.nan, inplace = True)
    log_dm1.dropna(inplace = True)
    Xp = log_dm1[["CT.SIZE", "CT.AVGINCOME", "CT.ROOMS", "CT.OWN", "CT.DEGREE", "CT.VEHICLES"]]
    Xp1 = sm.add_constant(Xp)
    pred_dm1[log.columns[i]] = np.array(model.predict(Xp1))
```

```python
result1 = pred_dm1
for i in pred_dm1.columns:
    result1[i] = np.exp(pred_dm1[i])
```

```python
result1["CT.Geo_GEOID"] = log_dm1["CT.Geo_GEOID"]
result1["CT.Geo_NAME"] = log_dm1["CT.Geo_NAME"]
```

```python
cols = list(result1.columns)
cols = [cols[len(cols) - 2]] + [cols[len(cols) - 1]] + cols[:len(cols) - 2]
result1 = result1[cols]
```

```python
result1.head()
```

```python
result1.to_csv("pred_fmli_sf2015.csv")
```
