##### Coefficient

```python
import pandas as pd
import numpy as np
```

```python
real = pd.read_csv("sfbay_cex_1990_2015_2015USD.csv")
```

```python
real.dropna(inplace = True)
real.drop(real.columns[[0, 1]], inplace = True, axis = 1)
```

```python
pred = pd.read_csv("pred_5county_avg.csv")
pred.drop(pred.columns[[0, 1, 2]], inplace = True, axis = 1)
```

```python
coeff = pd.DataFrame(pred.values / real.values)
coeff.columns = real.columns
```

```python
coeff
```

##### 2015

```python
pred2015 = pd.read_csv("pred_fmli_sf2015.csv")
Geo_GEOID = pred2015["CT.Geo_GEOID"]
Geo_NAME = pred2015["CT.Geo_NAME"]
```

```python
pred2015.drop(pred2015.columns[[0, 1, 2]], inplace = True, axis = 1)
```

```python
real2015 = pred2015.div(np.array(coeff.loc[0, :]), axis = "columns")
```

```python
real2015["Geo_GEOID"] = Geo_GEOID
real2015["Geo_NAME"] = Geo_NAME
```

```python
cols = list(real2015.columns)
cols = [cols[len(cols) - 2]] + [cols[len(cols) - 1]] + cols[:len(cols) - 2]
real2015 = real2015[cols]
```

```python
real2015.head()
```

```python
real2015.to_csv("real_fmli_sf2015.csv")
```
