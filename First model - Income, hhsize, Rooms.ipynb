```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fmli = pd.read_csv("fmli_15_clean.csv")
```

```python
# Log every variables, save it in a new data frame for future use.
log = np.log(fmli)
log["WEIGHT"] = fmli["WEIGHT"]
```

##### Run linear regression model for every expenditure on HHSIZE

```python
coeff_hhsize = []
R2_hhsize = []
N_hhsize = []
inter_hhsize = []
for i in range(17, len(log.columns)):
    df = log[["HHSIZE", log.columns[i], "WEIGHT"]]
    df.replace(np.Inf, np.nan, inplace = True)
    df.dropna(inplace = True)
    X = df.iloc[:, 0]
    y = df.iloc[:, 1]
    model = LinearRegression().fit(np.array(X).reshape(-1, 1), np.array(y))
    coeff_hhsize.append(model.coef_[0])
    y_pred = model.predict(np.array(X).reshape(-1, 1))
    R2_hhsize.append(np.sum((y_pred - np.average(y)) ** 2) / np.sum((y_pred - y) ** 2))
    N_hhsize.append(len(y))
    inter_hhsize.append(model.intercept_)
```

##### Run linear regression model for every expenditure on INCOME

```python
coeff_income = []
R2_income = []
N_income = []
inter_income = []
for i in range(17, len(log.columns)):
    df = log[["INCOME", log.columns[i]]]
    df.replace(np.Inf, np.nan, inplace = True)
    df.replace(-np.Inf, np.nan, inplace = True)
    df.dropna(inplace = True)
    X = df.iloc[:, 0]
    y = df.iloc[:, 1]
    model = LinearRegression().fit(np.array(X).reshape(-1, 1), np.array(y))
    coeff_income.append(model.coef_[0])
    y_pred = model.predict(np.array(X).reshape(-1, 1))
    R2_income.append(np.sum((y_pred - np.average(y)) ** 2) / np.sum((y_pred - y) ** 2))
    N_income.append(len(y))
    inter_income.append(model.intercept_)
```

##### Run linear regression model for every expenditure on ROOM

```python
coeff_room = []
R2_room = []
N_room = []
inter_room = []
for i in range(17, len(log.columns)):
    df = log[["ROOMS", log.columns[i]]]
    df.replace(np.Inf, np.nan, inplace = True)
    df.replace(-np.Inf, np.nan, inplace = True)
    df.dropna(inplace = True)
    X = df.iloc[:, 0]
    y = df.iloc[:, 1]
    model = LinearRegression().fit(np.array(X).reshape(-1, 1), np.array(y))
    coeff_room.append(model.coef_[0])
    y_pred = model.predict(np.array(X).reshape(-1, 1))
    R2_room.append(np.sum((y_pred - np.average(y)) ** 2) / np.sum((y_pred - y) ** 2))
    N_room.append(len(y))
    inter_room.append(model.intercept_)

R2_room
```

##### First Table: Elasticity

```python
index = []
for i in range(len(log.columns[17: len(log.columns)].values)):
    index.append(log.columns[17: len(log.columns)].values[i])
for i in range(len(log.columns[17: len(log.columns)].values)):
    index.append(log.columns[17: len(log.columns)].values[i])
```

```python
elasticity = pd.DataFrame({"HHSIZE": np.concatenate([coeff_hhsize, inter_hhsize]),
              "INCOME": np.concatenate([coeff_income, inter_income]),
              "ROOMS": np.concatenate([coeff_room, inter_room]),
              "category": ["Coefficient"] * 53 + ["Intercept"] * 53,
              "index":  index})
elasticity = pd.DataFrame(elasticity.groupby(["index", "category", "HHSIZE", "INCOME", "ROOMS"]).apply(list)).drop([0], axis = 1)

with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(elasticity)
```

##### Second Table: R_square and N

```python
R2N = pd.DataFrame({"HHSIZE": np.concatenate([R2_hhsize, N_hhsize]),
                    "INCOME": np.concatenate([R2_income, N_income]),
                    "ROOMS": np.concatenate([R2_room, N_room]),
                    "category": ["R_square"] * 53 + ["N"] * 53,
                    "index":  index})
R2N = pd.DataFrame(R2N.groupby(["index", "category", "HHSIZE", "INCOME", "ROOMS"]).apply(list)).drop([0], axis = 1)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(R2N)
```
