##### Run linear regression model for every expenditure on HHSIZE

```python
log.columns[17]
```

```python
coeff_hhsize = []
R2_hhsize = []
R2_adjust_hhsize = []
N_hhsize = []
inter_hhsize = []
p_hhsize = []
for i in range(17, len(log.columns)):
    df = log[["HHSIZE", log.columns[i]]]
    df.replace(np.Inf, np.nan, inplace = True)
    df.dropna(inplace = True)
    X = df[["HHSIZE"]]
    y = df.iloc[:, 1]
    #model = LinearRegression().fit(X, y)
    #coeff_hhsize.append(model.coef_[0])
    #y_pred = model.predict(X)
    #R2_adjust_hhsize.append(1 - (1 - model.score(X, y)) * (len(y) - 1)/(len(y) - X.shape[1] - 1))
    #R2_hhsize.append(model.score(X, y))
    N_hhsize.append(len(y))
    #inter_hhsize.append(model.intercept_)
    X1 = sm.add_constant(X)
    model = sm.OLS(y, X1).fit()
    #pred = model.predict(X1)
    #R2_hhsize.append(model.rsquared)
    R2_adjust_hhsize.append(model.rsquared_adj)
    coeff_hhsize.append(model.params[1])
    p_hhsize.append(model.pvalues[1])
```

##### Run linear regression model for every expenditure on INCOME

```python
coeff_income = []
R2_income = []
R2_adjust_income = []
N_income = []
inter_income = []
p_income = []
for i in range(17, len(log.columns)):
    df = log[["INCOME", log.columns[i]]]
    df.replace(np.Inf, np.nan, inplace = True)
    df.replace(-np.Inf, np.nan, inplace = True)
    df.dropna(inplace = True)
    X = df.iloc[:, 0]
    y = df.iloc[:, 1]
    #model = LinearRegression().fit(np.array(X).reshape(-1, 1), np.array(y))
    #coeff_income.append(model.coef_[0])
    #y_pred = model.predict(np.array(X).reshape(-1, 1))
    #R2_income.append(np.sum((y_pred - np.average(y)) ** 2) / np.sum((y_pred - y) ** 2))
    N_income.append(len(y))
    #inter_income.append(model.intercept_)
    X1 = sm.add_constant(X)
    model = sm.OLS(y, X1).fit()
    #pred = model.predict(X1)
    #R2_income.append(model.rsquared)
    R2_adjust_income.append(model.rsquared_adj)
    coeff_income.append(model.params[1])
    p_income.append(model.pvalues[1])
```

##### Run linear regression model for every expenditure on ROOM

```python
coeff_room = []
R2_room = []
R2_adjust_room = []
N_room = []
inter_room = []
p_room = []
for i in range(17, len(log.columns)):
    df = log[["ROOMS", log.columns[i]]]
    df.replace(np.Inf, np.nan, inplace = True)
    df.replace(-np.Inf, np.nan, inplace = True)
    df.dropna(inplace = True)
    X = df.iloc[:, 0]
    y = df.iloc[:, 1]
    #model = LinearRegression().fit(np.array(X).reshape(-1, 1), np.array(y))
    #coeff_room.append(model.coef_[0])
    #y_pred = model.predict(np.array(X).reshape(-1, 1))
    #R2_room.append(np.sum((y_pred - np.average(y)) ** 2) / np.sum((y_pred - y) ** 2))
    N_room.append(len(y))
    #inter_room.append(model.intercept_)
    X1 = sm.add_constant(X)
    model = sm.OLS(y, X1).fit()
    #pred = model.predict(X1)
    #R2_room.append(model.rsquared)
    R2_adjust_room.append(model.rsquared_adj)
    coeff_room.append(model.params[1])
    p_room.append(model.pvalues[1])
```

##### Multivariable Model: HHSIZE + INCOME + ROOMS

```python
N = []
adj_r = []
b_hhsize = []
b_income = []
b_rooms = []
se_hhsize = []
se_income = []
se_rooms = []
p_value = []
pred_y = pd.DataFrame()

for i in range(17, len(log.columns)):
    df = log[["HHSIZE", "INCOME", "ROOMS", log.columns[i]]]
    df.replace(np.Inf, np.nan, inplace = True)
    df.replace(-np.Inf, np.nan, inplace = True)
    df.dropna(inplace = True)
    df.reset_index(inplace = True)
    X = df[["HHSIZE", "INCOME", "ROOMS"]]
    Y = df[log.columns[i]]
    X1 = sm.add_constant(X)
    sf1 = sm.add_constant(log_X)
    model = sm.OLS(Y, X1).fit()
    pred_y[log.columns[i]] = np.array(model.predict(sf1))
    N.append(len(Y))
    b_hhsize.append(model.params[1])
    b_income.append(model.params[2])
    b_rooms.append(model.params[3])
    adj_r.append(model.rsquared_adj)
    se_hhsize.append(model.bse[1])
    se_income.append(model.bse[2])
    se_rooms.append(model.bse[3])
    p_value.append(model.t_test(model.params).pvalue)
```

```python
p_value = np.reshape(p_value, (1, 53))[0]
```

```python
b_hhsize = [round(x, 3) for x in b_hhsize]
b_income = [round(x, 3) for x in b_income]
b_rooms = [round(x, 3) for x in b_rooms]
se_hhsize = [round(x, 3) for x in se_hhsize]
se_income = [round(x, 3) for x in se_income]
se_rooms = [round(x, 3) for x in se_rooms]
N = [round(x, 0) for x in N]
adj_r = [round(x, 3) for x in adj_r]
p_value = [round(x, 5) for x in p_value]
```

```python
parameter = pd.DataFrame({"parameter": ["Beta"] * 53 + ["SE"] * 53,
                          "HHSIZE": np.concatenate([b_hhsize, se_hhsize]),
                          "INCOME": np.concatenate([b_income, se_income]),
                          "ROOMS": np.concatenate([b_rooms, se_rooms]),
                          "N": N + [" "] * 53,
                          "adj_r2": adj_r + [" "] * 53,
                          "p value": p_value + [" "] * 53,
                          "index": index[1:107]})
```

```python
parameter = pd.DataFrame(parameter.groupby(["index", "N", "adj_r2", "p value", "parameter", "HHSIZE", "INCOME", "ROOMS"]).apply(list)).drop([0], axis = 1)
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(parameter)
```
