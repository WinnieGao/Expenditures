```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy import stats
```

```python
fmli = pd.read_csv("fmli_15_clean.csv")
```

```python
# Log every variables, save it in a new data frame for future use.
log = np.log(fmli)
log["DETTACHED"] = fmli["DETTACHED"]
log["OWN"] = fmli["OWN"]
log["DEGREE"] = fmli["DEGREE"]
```

```python
log.columns
```

##### Multivariable Model: HHSIZE + INCOME + ROOMS + DETTACHED + OWN + DEGREE + VEH

```python
N = []
adj_r = []
b_hhsize = []
b_income = []
b_rooms = []
b_det = []
b_own = []
b_degree = []
b_veh = []
se_hhsize = []
se_income = []
se_rooms = []
se_det = []
se_own = []
se_degree = []
se_veh = []
p_hhsize = []
p_income = []
p_rooms = []
p_det = []
p_own = []
p_degree = []
p_veh = []

for i in range(17, len(log.columns)):
    df = log[["HHSIZE", "INCOME", "ROOMS", "DETTACHED", "OWN", "DEGREE", "VEHICLES", log.columns[i]]]
    df.replace(np.Inf, np.nan, inplace = True)
    df.replace(-np.Inf, np.nan, inplace = True)
    df.dropna(inplace = True)
    df.reset_index(inplace = True)
    X = df[["HHSIZE", "INCOME", "ROOMS", "DETTACHED", "OWN", "DEGREE", "VEHICLES"]]
    Y = df[log.columns[i]]
    X1 = sm.add_constant(X)
    model = sm.OLS(Y, X1).fit()
    N.append(len(Y))
    b_hhsize.append(model.params[1])
    b_income.append(model.params[2])
    b_rooms.append(model.params[3])
    b_det.append(model.params[4])
    b_own.append(model.params[5])
    b_degree.append(model.params[6])
    b_veh.append(model.params[7])
    adj_r.append(model.rsquared_adj)
    se_hhsize.append(model.bse[1])
    se_income.append(model.bse[2])
    se_rooms.append(model.bse[3])
    se_det.append(model.bse[4])
    se_own.append(model.bse[5])
    se_degree.append(model.bse[6])
    se_veh.append(model.bse[7])
    p_hhsize.append(model.pvalues[1])
    p_income.append(model.pvalues[2])
    p_rooms.append(model.pvalues[3])
    p_det.append(model.pvalues[4])
    p_own.append(model.pvalues[5])
    p_degree.append(model.pvalues[6])
    p_veh.append(model.pvalues[7])
```

```python
b_hhsize = [round(x, 3) for x in b_hhsize]
b_income = [round(x, 3) for x in b_income]
b_rooms = [round(x, 3) for x in b_rooms]
b_det = [round(x, 3) for x in b_det]
b_own = [round(x, 3) for x in b_own]
b_degree = [round(x, 3) for x in b_degree]
b_veh = [round(x, 3) for x in b_veh]

se_hhsize = [round(x, 3) for x in se_hhsize]
se_income = [round(x, 3) for x in se_income]
se_rooms = [round(x, 3) for x in se_rooms]
se_det = [round(x, 3) for x in se_det]
se_own = [round(x, 3) for x in se_own]
se_degree = [round(x, 3) for x in se_degree]
se_veh = [round(x, 3) for x in se_veh]

p_hhsize = [round(x, 3) for x in p_hhsize]
p_income = [round(x, 3) for x in p_income]
p_rooms = [round(x, 3) for x in p_rooms]
p_det = [round(x, 3) for x in p_det]
p_own = [round(x, 3) for x in p_own]
p_degree = [round(x, 3) for x in p_degree]
p_veh = [round(x, 3) for x in p_veh]

N = [round(x, 0) for x in N]
adj_r = [round(x, 3) for x in adj_r]
```

```python
index = []
for i in range(len(log.columns[17: len(log.columns)].values)):
    index.append(log.columns[17: len(log.columns)].values[i])
for i in range(len(log.columns[17: len(log.columns)].values)):
    index.append(log.columns[17: len(log.columns)].values[i])
for i in range(len(log.columns[17: len(log.columns)].values)):
    index.append(log.columns[17: len(log.columns)].values[i])
```

```python
parameter = pd.DataFrame({"parameter": ["Beta"] * 53 + ["SE"] * 53 + ["Confidence Level"] * 53,
                          "HHSIZE": np.concatenate([b_hhsize, se_hhsize, p_hhsize]),
                          "INCOME": np.concatenate([b_income, se_income, p_income]),
                          "ROOMS": np.concatenate([b_rooms, se_rooms, p_rooms]),
                          "DETTACHED": np.concatenate([b_det, se_det, p_det]),
                          "OWN": np.concatenate([b_own, se_own, p_own]),
                          "DEGREE": np.concatenate([b_degree, se_degree, p_degree]),
                          "VEH": np.concatenate([b_veh, se_veh, p_veh]),
                          "N": N + [" "] * 53 + ["  "] * 53,
                          "adj_r2": adj_r + [" "] * 53 + ["  "] * 53,
                          "index": index})
```

```python
parameter = pd.DataFrame(parameter.groupby(["index", "N", "adj_r2", "parameter", "HHSIZE", "INCOME", "ROOMS", "DETTACHED", "OWN", "DEGREE", "VEH"]).apply(list)).drop([0], axis = 1)
```
