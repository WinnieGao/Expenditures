```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style = "whitegrid", 
        color_codes = True,
        font_scale = 1.5)
```

```python
fmli = pd.read_csv("fmli_15_clean.csv")
```

```python
plot = fmli.drop(['NEWID', 'AGE', 'DETTACHED', 'OWN', 'DEGREE', 'WHITE', 'NORTHEAST', 'MIDWEST', 'SOUTH', 'WEST', 'MALE', 'MSA'], axis=1)
```

Scatterplots for HHSIZE vs. All other Expenditures

```python
hhsize = plot.drop(['INCOME', 'WEIGHT', 'ROOMS', 'VEHICLES'], axis = 1)
```

```python
sns.set(font_scale = 1)
for i in range(1, 10):
    if i == 9:
        sns.pairplot(data = hhsize, y_vars=['HHSIZE'], x_vars=hhsize.columns[range(6 * i - 5, 6 * i)].values, plot_kws = {'alpha': 0.6, 's': 5, 'edgecolor': 'k'}, size = 4)
    else:
        sns.pairplot(data = hhsize, y_vars=['HHSIZE'], x_vars=hhsize.columns[range(6 * i - 5, 6 * i + 1)].values, plot_kws = {'alpha': 0.6, 's': 5, 'edgecolor': 'k'}, size = 4)
```
